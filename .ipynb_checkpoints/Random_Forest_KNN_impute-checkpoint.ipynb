```python
def output_test(rf,x,filename):
    y=np.array(rf.predict(x)).reshape(-1,1)
    hey=np.arange(892,1310,1).reshape(-1,1)
    filename=filename+".csv"
    hey=np.hstack((hey,y))
    np.savetxt(filename,hey,header="PassengerId,Survived",delimiter=',')
    
```

```python
#filling missing values in "Age" column of trainng set
listg=["Pclass","Sex","SibSp","Parch"]
tar=datatrain["Age"].copy()
att=datatrain[listg].copy()
datatrain["Age"]=knn_impute(target=tar,attributes=att,
                            k_neighbors=10,missing_neighbors_threshold=0.8,
                            aggregation_method="mean", numeric_distance='euclidean',
                            categorical_distance='jaccard')

#filling missing values in "Embarked" column of trainng set
listg=["Pclass","Sex","SibSp","Parch","Fare","Age"]
tar=datatrain["Embarked"].copy()
att=datatrain[listg].copy()
datatrain["Embarked"]=knn_impute(target=tar,attributes=att,
                                 k_neighbors=10,missing_neighbors_threshold=0.8,
                                 aggregation_method="mode", numeric_distance='euclidean',
                                 categorical_distance='jaccard')
```

```python
#filling missing values in "Age" column of testing set
listg=["Pclass","Sex","SibSp","Parch"]
tar=datatest["Age"].copy()
att=datatest[listg].copy()
datatest["Age"]=knn_impute(target=tar,attributes=att,
                            k_neighbors=10,missing_neighbors_threshold=0.8,
                            aggregation_method="mean", numeric_distance='euclidean',
                            categorical_distance='jaccard')

#filling missing values in "Embarked" column of trainng set
listg=["Pclass","Sex","SibSp","Parch","Embarked","Age"]
tar=datatest["Fare"].copy()
att=datatest[listg].copy()
datatest["Fare"]=knn_impute(target=tar,attributes=att,
                            k_neighbors=10,missing_neighbors_threshold=0.8,
                            aggregation_method="mean", numeric_distance='euclidean',
                            categorical_distance='jaccard')
```

```python
le=LabelEncoder()
datatrain["Sex"]=le.fit_transform(datatrain["Sex"])
datatrain["Embarked"]=le.fit_transform(datatrain["Embarked"])
datatest["Sex"]=le.fit_transform(datatest["Sex"])
datatest["Embarked"]=le.fit_transform(datatest["Embarked"])
datatrain=datatrain.astype(float)
datatest=datatest.astype(float)
```

Now we have filled missing places so now lets carry on with Random Forest

```python
#converting to numpy array
xtrain=np.array(datatrain.iloc[:,2:])
ytrain=np.array(datatrain.iloc[:,1]).reshape(-1,1)
xtest=np.array(datatest.iloc[:,1:])
print(xtrain.shape,ytrain.shape)
print(xtest.shape)
```

Now lets use RandomSearchCV in scikit library to get a rough idea of optimum parameters...further we will nurture them more

```python
rf=RandomForestClassifier(random_state=91)
print(rf.get_params)
```

```python
params_try={"n_estimators":[int(i) for i in np.linspace(10,200,30)],  
            "max_depth":[int(i) for i in np.linspace(10, 110,20)], 
            "min_samples_split":[2,5,10,15,20,25], 
            "min_samples_leaf":[int(i) for i in np.linspace(1,50,20)], 
            "max_features":['auto','sqrt'],  
            "bootstrap":[True,False],   
            }
```

```python
random_search=RandomizedSearchCV(rf,param_distributions=params_try,n_iter=500,cv=5,verbose=2,random_state=91,n_jobs=-1)
```

```python
random_search=random_search.fit(xtrain,ytrain.ravel())
```

```python
best_random_params=random_search.best_params_
print(rf)
```

```python
#features selected by random estimation
ran_n_estimators=best_random_params["n_estimators"]
ran_min_samples_split=best_random_params["min_samples_split"]
ran_min_samples_leaf=best_random_params["min_samples_leaf"]
ran_max_features=best_random_params["max_features"]
ran_max_depth=best_random_params["max_depth"]
ran_bootstrap=best_random_params["bootstrap"]
```

Now we have rough idea of features so now we can do 
GridSearch around these parameter values to find more better parameter values

```python
rf=RandomForestClassifier()
print(rf)
```

```python
params_try={"n_estimators":[int(i) for i in np.linspace(ran_n_estimators-10,ran_n_estimators+10,10)],  
            "max_depth":[int(i) for i in np.linspace(ran_max_depth-10,ran_max_depth+10,10)], 
            "min_samples_split":[int(i) for i in np.linspace(ran_min_samples_split-2,ran_min_samples_split+2,4)], 
            "min_samples_leaf":[int(i) for i in np.linspace(ran_min_samples_leaf,ran_min_samples_leaf+10,10)], 
            "max_features":[ran_max_features],
            "bootstrap":[ran_bootstrap]   
            }
```

```python
grid_search=GridSearchCV(rf,param_grid=params_try,cv=3,n_jobs=-1,verbose = 2)
grid_search.fit(xtrain,ytrain.ravel())
```

```python
grid_best_params=grid_search.best_params_
print(grid_best_params)
best_rf_grid=grid_search.best_estimator_
```

```python
output_test(best_rf_grid,xtest,"RF_submission")
```
